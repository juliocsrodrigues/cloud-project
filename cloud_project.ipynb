<strong><i>Com os comandos abaixo podemos escolher o parametro do filtro de busca por UF ou tecnologia.</i></strong>

```python
while True:
    
    opcao = input ('por qual coluna deseja efetuar o filtro? digite o numero correspondente para sua opção:\n1 - UF \n2 - Tecnologia \nSelecionado: ')

    if opcao == '1': 
        filtro_uf = input("Digite o UF do estado que deseja filtrar: ").upper()
        # uf significa unidade federativa, representação dos estados brasileiros
        result = df[df['sigla_uf'] == filtro_uf]
        
        break
    
    elif opcao == '2':
        while True:
            opcao_tec = input ("selecione o numero correspondente a tecnologia desejada:\n1 - Fibra\n2 - Rádio \n3 - Satélite \n Selicionado: ")
            if opcao_tec =='1':
                filtro_tec = 'Fibra' 
                result = df[df['tecnologia'] == filtro_tec]
                break
                
            elif opcao_tec =='2':
                filtro_tec = 'Rádio' 
                result = df[df['tecnologia'] == filtro_tec]
                break
                
            elif opcao_tec =='3':
                filtro_tec = 'Satélite' 
                result = df[df['tecnologia'] == filtro_tec]
                break
                
            else: 
                     print ("Opção invalida ") 
        break
    else: 
     print ("Opção invalida ") 



```
